```python
import pandas as pd

ethiopia = pd.read_csv("data/ethiopia_clean.csv")
kenya = pd.read_csv("data/kenya_clean.csv")
nigeria = pd.read_csv("data/nigeria_clean.csv")
sudan = pd.read_csv("data/sudan_clean.csv")
tanzania = pd.read_csv("data/tanzania_clean.csv")
```

```python
ethiopia["country"] = "Ethiopia"
kenya["country"] = "Kenya"
nigeria["country"] = "Nigeria"
sudan["country"] = "Sudan"
tanzania["country"] = "tanzania"
```

```python
df_all = pd.concat([ethiopia, kenya, nigeria, sudan, tanzania])
```

```python
import matplotlib.pyplot as plt

df_all.groupby(["country", "month"])["T2M"].mean().unstack(0).plot()

plt.title("Monthly Temperature Comparison")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
df_all.groupby(["country", "month"])["PRECTOTCORR"].mean().unstack(0).plot(kind="bar")

plt.title("Monthly Rainfall Comparison")
plt.ylabel("Rainfall")
plt.show()
```

```python
extreme_heat = df_all[df_all["T2M_MAX"] > 10].groupby("country").size()

if len(extreme_heat) > 0:
    extreme_heat.plot(kind="bar")
    plt.title("Extreme Heat Events by Country")
    plt.show()
else:
    print("No extreme heat events found (>35°C)")
plt.show()
```
