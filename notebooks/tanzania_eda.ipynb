```python
import pandas as pd
import numpy as np
df = pd.read_csv("data/tanzania.csv")
df.replace(-999, np.nan, inplace=True)
df.duplicated().sum()
df.describe()
df["date"] = pd.to_datetime(df["YEAR"] * 1000 + df["DOY"], format="%Y%j")
df["month"] = df["date"].dt.month
df[["YEAR", "DOY", "date", "month"]].head()
df.head()
```

```python
import matplotlib.pyplot as plt

df.groupby("month")["T2M"].mean().plot(marker="o")
plt.title("This graph shows monthly temperature changes")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
df.groupby("month")["PRECTOTCORR"].sum().plot(kind="bar")
plt.title("Monthly Rainfall Pattern")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm")
plt.title("Climate Variable Correlation")
plt.show()
```
